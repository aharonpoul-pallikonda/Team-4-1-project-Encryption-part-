```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding
from cryptography.hazmat.primitives import hmac
from cryptography.hazmat.primitives.kdf.hkdf import HKDF
from cryptography.hazmat.primitives.kdf.concatkdf import ConcatKDFHash
from base64 import urlsafe_b64encode, urlsafe_b64decode
import os

# Function to generate a random key for AES-256 encryption
def generate_key():
    return os.urandom(32)  # 32 bytes for AES-256

# Function to encrypt the text file
def encrypt_file(file_path, key):
    # Read the contents of the file
    with open(file_path, 'rb') as f:
        plaintext = f.read()

    # Generate a random 96-bit (12 bytes) IV for GCM
    iv = os.urandom(12)

    # Create the AES-256 GCM cipher
    encryptor = Cipher(
        algorithms.AES(key),
        modes.GCM(iv),
        backend=default_backend()
    ).encryptor()

    # Encrypt the plaintext and get the associated tag
    ciphertext = encryptor.update(plaintext) + encryptor.finalize()
    tag = encryptor.tag

    # Write the IV, tag, and ciphertext to a new file
    with open(file_path + '.enc', 'wb') as f:
        f.write(iv + tag + ciphertext)

    print(f"File '{file_path}' has been encrypted and saved as '{file_path}.enc'")

# Main function to encrypt the file
def main():
    # Path to the text file to be encrypted
    file_path = '/content/drive/MyDrive/sample3.txt'  # Replace with your file path

    # Generate a key for AES-256 encryption (or you can securely store/use a predefined key)
    key = generate_key()

    # Encrypt the file
    encrypt_file(file_path, key)

    # Print the key (In practice, you should securely store the key instead of printing it)
    print(f"Encryption key (base64): {urlsafe_b64encode(key).decode()}")

if __name__ == '__main__':
    main()
```
